```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
pd.options.display.float_format = '{:,.2f}'.format

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
data = pd.read_csv('cost_revenue_dirty.csv')
```

```python
# Number of rows and columns
data.shape
```

```python
# Find any Nan values
print(f'Any Nan values in the data?\n{data.isna().values.any()}')
```

```python
# Check for null values
data.info()
```

```python
# Check for duplicated rows
print(f'Any duplicate rows?\n{data.duplicated().any()}')
```

```python
# Default data types of columns
data.dtypes
```

```python
# Convert objects to strings and then numerics
# Have to remove all non-numeric characters
chars_to_remove = ['$', ',']
clean_columns = ['USD_Production_Budget',
                'USD_Worldwide_Gross',
                'USD_Domestic_Gross']

for col in clean_columns:
    for char in chars_to_remove:
        data[col] = data[col].astype(str).str.replace(char, "")
    data[col] = pd.to_numeric(data[col])
```

```python
# Convert Release_Date to pandas Datetime type
data.Release_Date = pd.to_datetime(data.Release_Date)
data.head()
```

```python
data.info()
```

```python
# Full description of Dataframe
data.describe()
```

```python
# Lowest budget movie
data[data.USD_Production_Budget == 1100.00]
```

```python
# Highest budget movie
data[data.USD_Production_Budget == 425000000.00]
```

```python
# Movies that grossed $0 dollars domestically
domestic = data[data.USD_Domestic_Gross == 0.00]
print(f'Number of movies that grossed $0 domestically\n{len(domestic)}')
domestic.sort_values('USD_Production_Budget', ascending=False)
```

```python
# Movies that grossed $0 dollars domestically
worldwide = data[data.USD_Worldwide_Gross == 0.00]
print(f'Number of movies that grossed $0 worldwide\n{len(worldwide)}')
domestic.sort_values('USD_Production_Budget', ascending=False)
```

```python
# A subset for Worldwide releases that made
# zero revenue in the US
worldwide_releases = data.query('USD_Domestic_Gross == 0 and USD_Worldwide_Gross !=0')
print(f'Number of worldwide releases: {len(worldwide_releases)}')
worldwide_releases.tail()
```

```python
# Date of Data Collection
scrape_date = pd.Timestamp('2018-5-1')
```

```python
future_releases = data[data.Release_Date >= scrape_date]
print(f'Number of unreleased movies: {len(future_releases)}')
future_releases
```

```python
# New dataframe that does not
# include the unreleased movies
clean_data = data.drop(future_releases.index)
```

```python
# Movies that lost money
lost_money = clean_data.query('USD_Production_Budget > USD_Worldwide_Gross')
print(f'Number of movies that lost money: {len(lost_money)}')
```

```python
# Create a bubble chart using Seaborn
# to visualise the data
plt.figure(figsize=(8,4), dpi=200)

with sns.axes_style("darkgrid"):
    ax = sns.scatterplot(data=clean_data, 
                    x='Release_Date', 
                    y='USD_Production_Budget',
                    hue='USD_Worldwide_Gross',
                    size='USD_Worldwide_Gross',)

    ax.set(ylim=(0, 450000000),
           xlim=(clean_data.Release_Date.min(), clean_data.Release_Date.max()),
           xlabel='Year',
           ylabel='Budget in $100 millions')
```

```python
dt_index = pd.DatetimeIndex(clean_data.Release_Date)
years = dt_index.year
```

```python
# How to convert the year 1999 to the 90s decade
1999//10
```

```python
199*10
```

```python
decades = years//10*10
clean_data['Decade'] = decades
```

```python
# Create two new Dataframes
old_movies = clean_data[clean_data.Decade <= 1960]
new_movies = clean_data[clean_data.Decade > 1960]
```

```python
# Number of movies released before 1970
old_movies.describe()
```

```python
# Most expensive movie before 1970
old_movies.sort_values('USD_Production_Budget', ascending=False).head()
```

```python
# Create a scatter plot using the old_movies Dataframe
# Use linear regression to visuuaise the relationship
# between movies budget and worldwide revenue
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
  sns.regplot(data=old_movies, 
            x='USD_Production_Budget', 
            y='USD_Worldwide_Gross',
            scatter_kws = {'alpha': 0.4},
            line_kws = {'color': 'black'})
```

```python
# Create a scatter plot using the new_movies Dataframe
# Use linear regression to visuuaise the relationship
# between movies budget and worldwide revenue
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style('darkgrid'):
  ax = sns.regplot(data=new_movies,
                   x='USD_Production_Budget',
                   y='USD_Worldwide_Gross',
                   color='#2f4b7c',
                   scatter_kws = {'alpha': 0.3},
                   line_kws = {'color': '#ff7c43'})
  
  ax.set(ylim=(0, 3000000000),
         xlim=(0, 450000000),
         ylabel='Revenue in $ billions',
         xlabel='Budget in $100 millions')
```

```python
# Create regression object
regression = LinearRegression()
```

```python
# Explanatory Variable(s) or Feature(s)
X = pd.DataFrame(new_movies, columns=['USD_Production_Budget'])

# Response Variable or Target
y = pd.DataFrame(new_movies, columns=['USD_Worldwide_Gross'])
```

```python
# Find the best-fit line
regression.fit(X, y)
```

```python
# Theta zero
regression.intercept_
```

```python
# Theta one
regression.coef_
```

```python
# R-squared
regression.score(X, y)
```

```python
# Created a linear expression for the 
# old_movies dataframe
X = pd.DataFrame(old_movies, columns=['USD_Production_Budget'])
y = pd.DataFrame(old_movies, columns=['USD_Worldwide_Gross'])
regression.fit(X, y)
print(f'The slope coefficient is: {regression.coef_[0]}')
print(f'The intercept is: {regression.intercept_[0]}')
print(f'The r-squared is: {regression.score(X, y)}')
```

```python
# Use the created model to make a
# prediction for worldwide revenue
22821538 + 1.64771314 * 350000000
```

```python
# Estimated revenue for the movie
budget = 350000000
revenue_estimate = regression.intercept_[0] + regression.coef_[0,0]*budget
revenue_estimate = round(revenue_estimate, -6)
print(f'The estimated revenue for a $350 movie is around ${revenue_estimate:.10}.')
```

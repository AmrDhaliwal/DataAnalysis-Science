```python
# Convert entire DATE column to Datetime
df.DATE = pd.to_datetime(df.DATE)
df.head()
```

```python
# Create a pivoted DataFrame

pivoted_df = df.pivot(index='DATE', columns='TAG', values='POSTS')
```

```python
pivoted_df.count()
```

```python
# Fill Nan values with 0
pivoted_df.fillna(0, inplace=True)
```

```python
pivoted_df.head()
```

```python
# Check to see if any Nan values
# left in pivoted_df

pivoted_df.isna().values.any()
```

```python
# A chart using matplotlib for java
# resize the entire chart

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
plt.plot(pivoted_df.index, pivoted_df['java'])
```

```python
# A chart using matplotlib to plot java and python

plt.figure(figsize=(16,10)) # make chart larger
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

plt.plot(pivoted_df.index, pivoted_df.java)
plt.plot(pivoted_df.index, pivoted_df.python)
```

```python
# A chart displaying all programming languages

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

# Use a for loop to plot all languages

for column in pivoted_df.columns:
    plt.plot(pivoted_df.index, pivoted_df[column], 
             linewidth=3, label=pivoted_df[column].name)

plt.legend(fontsize=16)
```

```python
# Smoothing out Time Series Data
# The window is number of observations that are averaged
roll_df = pivoted_df.rolling(window=3).mean()

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

# plot the roll_df instead
for column in roll_df.columns:
    plt.plot(roll_df.index, roll_df[column], 
             linewidth=3, label=roll_df[column].name)

plt.legend(fontsize=16)
```
